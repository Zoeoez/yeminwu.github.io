### Question 1

Do you want the user to be able to pick any 2 columns?  Or only certain columns?  Why or why not?


I do not want the user to be able to pick any 2 columns because some columns do not make sense together such as state and city. So I separete numerical and categorical variables. I first select ufos_numerical which are numerical variables because I want to use these as my colormap.I use ipywidgets to make my plot become interactive.


```python
ufos_numerical = ufos[["duration_seconds", "latitude", "longitude"]]
ufos_numerical
```

```python
@ipywidgets.interact(col_name1 = ufos_col.columns, col_name2 = ufos_col.columns, colormap = ufos_numerical.columns)
def create_plot(col_name1, col_name2, colormap):
    fig,ax = plt.subplots(figsize=(10,6))
    ax.scatter(ufos[col_name1], ufos[col_name2], c=np.log10(ufos[colormap]))
    ax.set_xlabel(col_name1)
    ax.set_ylabel(col_name2)
    plt.show()
```

### Question 2

As for question 2, I first select categorical variables. I use @ipywidgets.interact(column = ufos_categorical.columns, bins = bins_itext) to achieve the interactive function here. 

```python
ufos_categorical = ufos[["date", "city", "state", "country", "shape", "report_date"]]
ufos_categorical
```

```python
bins_itext = ipywidgets.IntText()
bins_itext.value = 20
@ipywidgets.interact(column = ufos_categorical.columns, bins = bins_itext)
def creat_hist(column, bins):
    fig,ax = plt.subplots(figsize=(8,6))
    ax.hist(x = ufos[column], bins = bins)
    ax.set_xlabel(column)
    plt.show()
```
