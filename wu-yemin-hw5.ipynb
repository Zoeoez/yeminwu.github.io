### bq plot


```python
!pip install bqplot
import bqplot
import traitlets
import ipywidgets
%matplotlib inline
import numpy as np
```

```python
x = np.random.random(100)
y = np.random.random(100)
```

```python
x_sc = bqplot.LinearScale()
y_sc = bqplot.LinearScale()
x_ax = bqplot.Axis(scale = x_sc, label = 'X Value')
y_ax = bqplot.Axis(scale = y_sc, label = 'Y Value', orientation = 'vertical')
```

```python
scatters=bqplot.Scatter(x = x, y = y, scales = {'x': x_sc, 'y': y_sc})
fig = bqplot.Figure(marks = [scatters], axes = [x_ax, y_ax],title="scatter plot using bqplot")
```
